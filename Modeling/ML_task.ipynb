# Import & Read file

```python
import pandas as pd
import numpy as np
import ML_models as ml
from sklearn import ensemble, preprocessing, metrics
from joblib import dump,load
```

```python
data = pd.read_csv('GraphTheoryDataset.csv',index_col=0)
data
```

```python
len(data[data.label == 0]), len(data[data.label != 0])
```

# Preprocessing

```python
Feature=np.array(data.iloc[:,1:-2])
Label=np.array(data.iloc[:,-2])
Feature.shape,Label.shape
```

```python
scaler = preprocessing.MinMaxScaler().fit(Feature)
Feature=scaler.transform(Feature)
```

# Malware Detection (Binary Classification)

```python
md_model = dict()

# let the label be binary
for i in range(len(Label)):
    if Label[i] != 0:
        Label[i] = 1
```

```python
md_model['rf'],tmp = ml.RF(Feature,Label,100)
```

```python
md_model['knn'],tmp = ml.KNN(Feature,Label,1)
```

```python
md_model['svm'],tmp = ml.SVM(Feature,Label,100,0.0001)
```

```python
md_model['mlp'],tmp = ml.MLP(Feature,Label,100,300)
```

# Family Classification (Multi-class Classification)

```python
fc_model = dict()

Label = data.label
```

```python
fc_model['rf'],tmp = ml.RF(Feature,Label,100)
```

```python
fc_model['knn'],tmp = ml.KNN(Feature,Label,1)
```

```python
fc_model['svm'],tmp = ml.SVM(Feature,Label,100,0.0001)
```

```python
fc_model['mlp'],tmp = ml.MLP(Feature,Label,100,300)
```

# Save model

```python
for k,m in md_model.items():
    dump(m,'../MD_Model/' + k + '.joblib')
for k,m in fc_model.items():
    dump(m,'../FC_Model/' + k + '.joblib')
```
